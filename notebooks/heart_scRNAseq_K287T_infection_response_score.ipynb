```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import plotnine as pn
import seaborn as sns
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, dpi_save= 300, fontsize=6, frameon=True, figsize=(2, 2), transparent=True, format="pdf")
```

```python
heart_infection_sc_K287T = sc.read_h5ad(filename="./../../../pyobjs_github/heart_infection_sc_K287T.h5ad")
heart_infection_sc_K287T
```

```python
sc.pl.umap(heart_infection_sc_K287T, color = ['celltypes_leiden_0.5'], legend_fontsize=10)
```

```python
day4_WT_infection_markers = pd.read_csv("./../../../csvs_github/heart_IFN_genelist_D4PI.csv", index_col=0)
day4_WT_infection_markers_significant = day4_WT_infection_markers[day4_WT_infection_markers["logfoldchanges"] > 1.0]
day4_WT_infection_markers_significant = day4_WT_infection_markers_significant[day4_WT_infection_markers_significant["pvals_adj"] < 0.01]
print(day4_WT_infection_markers_significant)
```

```python
sc.tl.score_genes(heart_infection_sc_K287T, day4_WT_infection_markers_significant["names"], use_raw=True, score_name = 'Infection_response_score')
```

```python
sc.settings.set_figure_params(dpi_save= 400, fontsize=6, facecolor='white', frameon=True, transparent=True, format="pdf")
fig, ax = plt.subplots(figsize=(1.5, 1.4))
plt.grid(b=None)
plt.rc('font', size=6)
ax.set_axisbelow(True)
sns.boxplot(data = heart_infection_sc_K287T[heart_infection_sc_K287T.obs['celltype_groups'].isin(["Cardiomyocytes", "Fibroblasts", "Endothelial cells", "Endocardial cells", "Smooth muscle cells", "Pericytes"]) & heart_infection_sc_K287T.obs['stage'].isin(["D4PI"])].obs, y= "Infection_response_score", x = "condition", linewidth = 0.5, fliersize = 0.5, hue = "celltype_groups", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='right')
plt.legend(ncol=1, prop={'size': 6}, handletextpad=0.5, markerscale = 0.6, frameon = False, labelspacing = 0.2, bbox_to_anchor = (1.0,1.0))
plt.savefig("figures/IFN_cardiac_allsamples.pdf", facecolor='w', edgecolor='w',
        orientation='potrait', format="pdf",
        transparent=True, metadata=None)
```

```python
sc.settings.set_figure_params(dpi_save= 400, fontsize=6, facecolor='white', frameon=True, transparent=True, format="pdf")
fig, ax = plt.subplots(figsize=(5, 1.4))
plt.grid(b=None)
plt.rc('font', size=6)
ax.set_axisbelow(True)
sns.boxplot(data = heart_infection_sc_K287T[heart_infection_sc_K287T.obs['celltype_groups'].isin(["Cardiomyocytes", "Fibroblasts", "Endothelial cells", "Endocardial cells",  "Smooth muscle cells", "Pericytes"]) & heart_infection_sc_K287T.obs['stage'].isin(["D4PI", "D7PI", "D10PI"])].obs, y= "Infection_response_score", x = "condition", linewidth = 0.5, fliersize = 0.5, hue = "celltype_groups", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='right')
plt.legend(ncol=1, prop={'size': 6}, handletextpad=0.5, markerscale = 0.6, frameon = False, labelspacing = 0.2, bbox_to_anchor = (1.0,1.0))
plt.show()
```
