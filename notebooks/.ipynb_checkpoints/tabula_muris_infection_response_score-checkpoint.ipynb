```python
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, dpi_save= 1000, facecolor='white', frameon=True, figsize=(5, 5))
```

```python
tabula_muris_droplet = sc.read_h5ad(filename="./../tabula-muris/00_data_ingest/tabula_muris_droplet_raw.h5ad")
```

```python
tabula_muris_FACS = sc.read_h5ad(filename="./../tabula-muris/00_data_ingest/tabula_muris_facs_raw.h5ad")
```

```python
tabula_muris = tabula_muris_droplet.concatenate(tabula_muris_FACS, 
                                                    batch_key= "modality", batch_categories=["droplet", "FACS"],
                                                    index_unique="-")
```

```python
tabula_muris.write_h5ad("./../tabula-muris/00_data_ingest/tabula_muris_merged_raw.h5ad")
```

```python
tabula_muris.var['mt'] = tabula_muris.var_names.str.startswith('Mt-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(tabula_muris, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(tabula_muris, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True, stripplot=True, groupby = "tissue", rotation= 90)
```

```python
sc.pp.normalize_total(tabula_muris, target_sum=1e4)
sc.pp.log1p(tabula_muris)
sc.pp.highly_variable_genes(tabula_muris, max_mean=4, min_disp=0.5)
sc.pl.highly_variable_genes(tabula_muris, )
# sc.tl.score_genes_cell_cycle(tabula_muris, s_genes=s_genes, g2m_genes=g2m_genes)
tabula_muris.raw = tabula_muris
tabula_muris = tabula_muris[:, tabula_muris.var.highly_variable]
# sc.pp.regress_out(tabula_muris, keys=['total_counts', 'pct_counts_mt', 'S_score', 'G2M_score'], n_jobs=20)
sc.pp.scale(tabula_muris, max_value=10)
sc.tl.pca(tabula_muris, svd_solver='arpack')
sc.pl.pca(tabula_muris, color = ["tissue"])
sc.pl.pca_variance_ratio(tabula_muris, n_pcs = 50, log=True)
sc.pp.neighbors(tabula_muris, n_pcs=20)
sc.tl.umap(tabula_muris)
```

```python
sc.settings.set_figure_params(dpi=80, dpi_save= 1000, facecolor='white', frameon=True, figsize=(5, 5))
sc.pl.umap(tabula_muris, color = ["tissue"])
```

```python
sc.settings.set_figure_params(dpi=80, dpi_save= 1000, facecolor='white', frameon=True, figsize=(3, 3))
sc.pl.umap(tabula_muris, color = ["cell_ontology_class"], save="tm_celltype_class")
```

```python
tabula_muris = sc.read_h5ad(filename="./../tabula-muris/00_data_ingest/tabula_muris_processed.h5ad")
```

```python
tabula_muris.obs["tissue1"] = tabula_muris.obs["tissue"]
celltypes = {'Aorta': 'Heart and Aorta',
             'Bladder': 'Bladder',
             'Brain_Myeloid': "Brain (Myeloid)",
             'Brain_Non-Myeloid': 'Brain (Non-myeloid)',
             'Diaphragm': 'Diaphragm',
             'Fat': 'Fat',
             'Heart': 'Heart and Aorta',
             'Heart_and_Aorta': 'Heart and Aorta',
             'Kidney': 'Kidney',
             'Large_Intestine': 'Large Intestine',
             'Limb_Muscle': 'Skeletal Muscle',
             'Liver': 'Liver',
             'Lung': 'Lung',
             'Mammary_Gland': 'Mammary Gland',
             'Marrow': 'Marrow',
             'Pancreas': 'Pancreas',
             'Skin': 'Skin',
             'Spleen': 'Spleen',
             'Thymus': 'Thymus',
             'Tongue': 'Tongue',
             'Trachea': 'Trachea'}
tabula_muris.obs["tissue1"] = tabula_muris.obs["tissue1"].map(celltypes).astype("category")
```

```python
np.unique(tabula_muris.obs[tabula_muris.obs["tissue1"] == "Brain (Myeloid)"]["cell_ontology_class"])
```

```python
IFN_gene_list = pd.read_csv("/fs/cbsuvlaminck3/workdir/mm2937/MouseInfection/csvs/heart_IFN_genelist_D4PI.csv", index_col=0)
IFN_gene_list = IFN_gene_list[(IFN_gene_list["logfoldchanges"] > 1.0) & (IFN_gene_list["pvals"] < 0.01)]
IFN_gene_list
```

```python
sc.tl.score_genes(tabula_muris, IFN_gene_list["names"], use_raw=True, score_name ='Infection_response_score')
```

```python
pw_adata_all = tabula_muris.obs
```

```python
pw_adata_all.loc[pw_adata_all["cell_ontology_class"].str.contains("endothelial"), "cell_ontology_group"] = "endothelial cell"
pw_adata_all.loc[pw_adata_all["cell_ontology_class"].str.contains("epithelial"), "cell_ontology_group"] = "epithelial cell"
pw_adata_all.loc[pw_adata_all["cell_ontology_class"].str.contains("smooth muscle"), "cell_ontology_group"] = "smooth muscle cell"
pw_adata_all.loc[pw_adata_all["cell_ontology_class"].str.contains("fibroblast"), "cell_ontology_group"] = "fibroblast cell"
pw_adata_all.loc[pw_adata_all["cell_ontology_class"].str.contains("mesenchymal"), "cell_ontology_group"] = "mesenchymal cell"
```

```python
pw_adata_all["cell_ontology_group"] = pw_adata_all["cell_ontology_group"].astype("string")
pw_adata_all["tissue1"] = pw_adata_all["tissue1"].astype("string")
```

```python
sc.settings.set_figure_params(dpi_save= 400, fontsize=6, facecolor='white', frameon=True, vector_friendly = False, transparent=True, format="pdf")
fig, ax = plt.subplots(figsize=(5, 1.8))
plt.rc('font', size=6)
plt.grid(b=None)
temp = pw_adata_all[pw_adata_all['cell_ontology_group'].isin(["endothelial cell", "fibroblast cell", "smooth muscle cell", "epithelial cell", "mesenchymal cell"])]
# temp = temp[~temp['tissue1'].isin(["Liver", "Pancreas", "Thymus", "Large Intestine"])]
sns.boxplot(data = temp, y= "Infection_response_score", x = "tissue1", hue = "cell_ontology_group", linewidth = 0.5, fliersize = 0.5, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
plt.legend(ncol=1, prop={'size': 6}, handletextpad=0.5, markerscale = 0.6, frameon = False, labelspacing = 0.2, bbox_to_anchor = (1.0,1.0))
plt.tight_layout()
plt.show()
```

```python
sc.settings.set_figure_params(dpi_save= 400, fontsize=6, facecolor='white', frameon=True, vector_friendly = False, transparent=True, format="pdf")
fig, ax = plt.subplots(figsize=(5, 1.8))
plt.rc('font', size=6)
plt.grid(b=None)
temp = pw_adata_all[pw_adata_all['cell_ontology_group'].isin(["endothelial cell", "fibroblast cell", "smooth muscle cell", "epithelial cell", "mesenchymal cell"])]
temp = temp[~temp['tissue1'].isin(["Liver", "Pancreas", "Thymus", "Large Intestine"])]
sns.boxplot(data = temp, y= "Infection_response_score", x = "tissue1", hue = "cell_ontology_group", linewidth = 0.5, fliersize = 0.5, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
plt.legend(ncol=1, prop={'size': 6}, handletextpad=0.5, markerscale = 0.6, frameon = False, labelspacing = 0.2, bbox_to_anchor = (1.0,1.0))
plt.tight_layout()
plt.savefig("/fs/cbsuvlaminck3/workdir/mm2937/MouseInfection/notebooks/figures/tabula-muris.pdf", facecolor='w', edgecolor='w',
        orientation='potrait', format="pdf",
        transparent=True, metadata=None)
```
